```python
# 구간별로 데이터를 나누고 누적 그래프 계산
bins = range(0, max(df['total_subs_num']) + 10, 10)
df['subs_group'] = pd.cut(df['total_subs_num'], bins=bins, right=False)
df['cumulative_subs'] = df.groupby('subs_group')['total_subs_num'].cumsum()

# 시각화
plt.figure(figsize=(10, 6))
for subs_group, group_data in df.groupby('subs_group'):
    plt.plot(group_data['total_subs_num'], group_data['cumulative_subs'], marker='o', label=subs_group)

plt.title('Cumulative Subscribers by 10s')
plt.xlabel('Total Subscribers')
plt.ylabel('Cumulative Subscribers')
plt.legend(title='Total Subscribers Group')
plt.grid(True)
plt.show()
```

```python
# total_subs_num이 100 미만인 경우에는 'LessThan100', 그렇지 않은 경우에는 '100OrMore'로 값을 할당하는 새로운 칼럼 생성
df['subs_status'] = np.where(df['total_subs_num'] < 100, 'LessThan100', '100OrMore')

# 결과 출력
print(df)
```

```python
# 시각화
plt.figure(figsize=(6, 4))
sns.barplot(x='subs_status', y='num_of_subs', data=df, palette='viridis')
plt.title('num_of_subs by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('num_of_subs')
plt.show()
```

```python
# 시각화
plt.figure(figsize=(6, 4))
sns.barplot(x='subs_status', y='total_community_posts_num', data=df, palette='viridis')
plt.title('total_community_posts_num by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('total_community_posts_num')
plt.show()
```

- 구독자가 100명이 안되는 집단은 커뮤니티에 거의 글을 쓰지 않는다. 
- 어쩌면 구독자가 없어서 글이 많이 올라오지 않는걸수도 있다. 
- 구독자의 수와 커뮤니티 글 개수 사이의 상관관계
