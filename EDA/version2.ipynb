```python
more_than_100_df['community_posts_ratio'] = more_than_100_df['total_community_posts_num'] / more_than_100_df['total_posts_num']
```

```python
more_than_100_df['community_posts_ratio'].value_counts()
```

```python
# 'community_posts_ratio'Í∞Ä 1ÏùÑ ÎÑòÎäî ÌñâÎì§
ratio_greater_than_1 = more_than_100_df[more_than_100_df['community_posts_ratio'] > 1]

# 'community_posts_ratio'Í∞Ä 1ÏùÑ ÎÑòÏßÄ ÏïäÎäî ÌñâÎì§
ratio_less_than_or_equal_to_1 = more_than_100_df[more_than_100_df['community_posts_ratio'] <= 1]

# Í∞ÅÍ∞ÅÏùò ÎπÑÏú® Í≥ÑÏÇ∞
ratio_greater_than_1_count = len(ratio_greater_than_1)
ratio_less_than_or_equal_to_1_count = len(ratio_less_than_or_equal_to_1)

# Ï†ÑÏ≤¥ ÌñâÏùò Í∞úÏàò
total_rows = len(more_than_100_df)

# ÎπÑÏú® Í≥ÑÏÇ∞
ratio_greater_than_1_percentage = (ratio_greater_than_1_count / total_rows) * 100
ratio_less_than_or_equal_to_1_percentage = (ratio_less_than_or_equal_to_1_count / total_rows) * 100

print(f'Ratio greater than 1: {ratio_greater_than_1_percentage:.2f}%') #Ìè¨Ïä§Ìä∏ Í∏ÄÎ≥¥Îã§ Ïª§ÎÆ§ÎãàÌã∞ Í∏ÄÏù¥ Îçî ÎßéÏùÄ ÌÅ¨Î¶¨ÏóêÏù¥ÌÑ∞
print(f'Ratio less than or equal to 1: {ratio_less_than_or_equal_to_1_percentage:.2f}%')
```
