```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/creator_community_2.csv')
df
```

```python
# subs_status가 'LessThan100'인 경우 필터링
less_than_100_df = df[df['subs_status'] == 'LessThan100']

# num_of_paid_subs / total_subs_num의 비율 계산
less_than_100_df['paid_subs_ratio'] = less_than_100_df['num_of__paid_subs'] / less_than_100_df['total_subs_num']
```

```python
more_than_100_df = df[df['subs_status'] == '100OrMore']

# num_of_paid_subs / total_subs_num의 비율 계산
more_than_100_df['paid_subs_ratio'] = more_than_100_df['num_of__paid_subs'] / more_than_100_df['total_subs_num']
```

```python
more_than_100_df['paid_subs_ratio'].hist()
```

```python
less_than_100_df['paid_subs_ratio'].isna().sum() / len(less_than_100_df['paid_subs_ratio'])
```
