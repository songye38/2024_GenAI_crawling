### 크리에이터가 몇개의 포스트를 작성했는지 크롤링하기 

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Chrome 브라우저 열기
driver = webdriver.Chrome()

# 웹 페이지로 이동
url = "https://steadio.co/creator/jwieut/posts"
driver.get(url)

# 페이지 로딩을 기다림 (필요한 경우)
driver.implicitly_wait(10)  # 예: 최대 10초까지 대기

# 페이지의 HTML 가져오기
html_content = driver.page_source

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, 'html.parser')

# 특정 요소 찾기
target_element = soup.find('div', class_='flex gap-[2px] font_title_bold_md content_secondary')

# 찾은 요소의 텍스트 출력
if target_element:
    print(target_element.span.text)
else:
    print("해당 요소를 찾을 수 없습니다.")

# 브라우저 닫기
driver.quit()
```

```python
# 함수 정의
def get_total_subs_num(url):
    url = "https://" + url.split("#")[0] + '/posts'
    # Chrome 브라우저 열기
    driver = webdriver.Chrome()

    # 웹 페이지로 이동
    driver.get(url)

    # 페이지 로딩을 기다림 (필요한 경우)
    driver.implicitly_wait(10)  # 예: 최대 10초까지 대기

    # 페이지의 HTML 가져오기
    html_content = driver.page_source

    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(html_content, 'html.parser')

    # strong class가 'font_label_bold_lg'인 요소 찾기
    strong_element = soup.find('div', class_='flex gap-[2px] font_title_bold_md content_secondary')

    

    # 찾은 요소의 텍스트 반환
    if strong_element:
        total_subs_num = strong_element.span.text
    else:
        total_subs_num = "해당 요소를 찾을 수 없습니다."

    # 브라우저 닫기
    driver.quit()

    return total_subs_num

# DataFrame의 'creator_url' 열에 함수 적용하여 'total_subs_num' 열에 결과 저장
df['total_posts_num'] = df['creator_url'].apply(get_total_subs_num)

# 결과 출력
print(df)
```
